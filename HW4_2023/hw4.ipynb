```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy

dataset1 = pd.read_csv("Q3/data1.tsv", sep='\t', index_col=0)
dataset2 = pd.read_csv("Q3/data2.tsv", sep='\t', index_col=0)

dataset2
```

```python
# Experiment 1 values
t_val, p_val = scipy.stats.ttest_rel(dataset1.iloc[:, 0], dataset1.iloc[:, 1])
print(f"ttest t-val: {t_val} and p-val: {p_val}")

stat_val, other_p_val = scipy.stats.wilcoxon(dataset1.iloc[:, 0], dataset1.iloc[:, 1])
print(f"wilcoxon stat-val: {stat_val} and p-val: {other_p_val}")
```

```python
# Experiment 2 values
t_val, p_val = scipy.stats.ttest_rel(dataset1.iloc[:, 2], dataset1.iloc[:, 3])
print(f"ttest t-val: {t_val} and p-val: {p_val}")

stat_val, other_p_val = scipy.stats.wilcoxon(dataset1.iloc[:, 2], dataset1.iloc[:, 3])
print(f"wilcoxon stat-val: {stat_val} and p-val: {other_p_val}")
```

```python
# Plots
fig, ax = plt.subplots(1,2,figsize=(8,4))
ax[0].scatter(dataset1.iloc[:, 0], dataset1.iloc[:, 1]) 
ax[0].set_title('Experiment 1')
ax[0].set_xlabel('Before treatment Expression Level (%)')
ax[0].set_ylabel('After treatment Expression Level (%)')
ax[1].scatter(dataset1.iloc[:, 2], dataset1.iloc[:, 3]) 
ax[1].set_title('Experiment 2')
ax[1].set_xlabel('Before treatment Expression Level (%)')
ax[1].set_ylabel('After treatment Expression Level (%)')
```
